```python
import pandas as pd
import numpy as np
import math
from scipy import stats
```

```python
def Z_and_p(x1, x2, sigma1, sigma2, n1, n2):
    z = (x1 - x2) / math.sqrt(((sigma1 ** 2) / n1) + ((sigma2 ** 2) / n2))
    
    if z < 0:
        p = stats.norm.cdf(z)
    else:
        p = 1 - stats.norm.cdf(z)
    
    print(z, p)
```

```python
# Function call
Z_and_p(121, 112, 8, 8, 10, 10)
```
