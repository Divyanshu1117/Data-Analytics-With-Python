```python
from scipy import stats
```

```python
b = [89.19, 90.95, 90.46, 93.21, 97.19, 97.04, 91.07, 92.75]
```

```python
a = [91.5, 94.18, 92.18, 95.39, 91.79, 89.07, 94.72, 89.21]
```

```python
# Two sample t-test (equal variance assumed)
result = stats.ttest_ind(a, b, equal_var=True)
print(result)
```

```python
# Critical t-value
critical_t = stats.t.ppf(0.025, 14)
print(critical_t)
```
