```python
import pandas as pd
import numpy as np
from scipy import stats
import math
```

```python
def samplesize(alfa,beta,mu1,mu2,sigma):
    z1 = -1*stats.norm.ppf(alfa)
    z2 = -1*stats.norm.ppf(beta)
    n = (((z1+z2)**2)*(sigma**2))/((mu1-mu2)**2)
    print(n)
```

```python
samplesize(0.05,0.1,12,12.75,3.2) # alpha=0.05 (Type I error), beta=0.1 (Type II error), mu1=12, mu2=12.75, sigma=3.2
```
