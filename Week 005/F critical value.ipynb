```python
import pandas as pd
import scipy.stats as stats
from statsmodels.stats.multicomp import MultiComparison
```

```python
f_critical = stats.f.ppf(1 - 0.05, dfn=4, dfd=20)
print(f_critical)
```

```python
df3 = pd.read_excel(r"cotton weight.xlsx")
df3.head()
```

```python
data1 = pd.melt(
    df3.reset_index(),
    id_vars=['index'],
    value_vars=['cotwt.15', 'cotwt.20', 'cotwt.25', 'cotwt.30', 'cotwt.35']
)

data1.columns = ['id', 'treatments', 'value']

data1.head()
```

```python
mc = MultiComparison(data1['value'], data1['treatments'])
mcresults = mc.tukeyhsd(0.05)
mcresults.summary()
```
