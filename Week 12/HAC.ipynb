```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
from scipy.cluster.hierarchy import fcluster
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist
```

```python
data = pd.read_excel("hierarchical_clustering.xlsx")
print(data)
```

```python
x = data['Variable 1']
y = data['Variable 2']
n = range(1, 8)

fig, ax = plt.subplots()
ax.scatter(x, y, marker='*', c='red', alpha=0.5)
plt.grid()
plt.xlabel("Variable 1")
plt.ylabel("Variable 2")
for i, txt in enumerate(n):
    ax.annotate(txt, (x[i], y[i]))
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

linked = linkage(data, 'single')

labelList = range(1, 8)

plt.figure(figsize=(10, 7))
dendrogram(linked,
           orientation='top',
           labels=labelList,
           distance_sort='descending',
           show_leaf_counts=True)

plt.axhline(y=2.5)
plt.show()
```

```python
import sklearn
from sklearn.cluster import AgglomerativeClustering

k = 2
Hclustering = AgglomerativeClustering(
    n_clusters=k,
    metric='euclidean',
    linkage='single'
)
Hclustering.fit(data)
```

```python
Hclustering.fit_predict(data)
```

```python
print(Hclustering.labels_)
```

```python
x = data['Variable 1']
y = data['Variable 2']
n = range(1, 8)

fig, ax = plt.subplots()
ax.scatter(x, y, c=Hclustering.labels_, cmap='rainbow')
plt.grid()
plt.xlabel("Variable 1")
plt.ylabel("Variable 2")
for i, txt in enumerate(n):
    ax.annotate(txt, (x[i], y[i]))
plt.show()
```
