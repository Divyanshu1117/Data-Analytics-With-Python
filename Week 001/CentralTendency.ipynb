```python
# Finding Minimum And Maximum Value:-
```

```python
data = [1, 3, 4, 463, 2, 3, 6, 8, 9, 4, 254, 6, 72]
min(data), max(data)
```

```python
data = [1, 3, 4, 463, 2, 3, 6, 8, 9, 4, 254, 6, 72]
def min_and_max(data):
    min_val = min(data)
    max_val = max(data)
    return (min_val, max_val)
```

```python
min_and_max(data)
```

```python
def rangef(data):
    min_val = min(data)
    max_val = max(data)
    return (max_val - min_val)
```

```python
rangef(data)
```
