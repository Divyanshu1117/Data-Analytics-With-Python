```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

```python
df = pd.read_excel('RBD2.xlsx')
df
```

```python
data = pd.melt(
    df.reset_index(),
    id_vars=['index'],
    value_vars=['chem1', 'chem2', 'chem3', 'chem4']
)
```

```python
data.columns = ['index', 'treatments', 'value']
```

```python
model = ols('value ~ C(treatments)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=1)
anova_table
```

```python
data = pd.melt(
    df.reset_index(),
    id_vars=['index'],
    value_vars=['chem1', 'chem2', 'chem3', 'chem4']
)
```

```python
data.columns = ['Fabric', 'Chemical', 'value']
data
```

```python
model = ols('value ~ C(Fabric) + C(Chemical)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=1)
anova_table
```
