```python
import pandas as pd
import numpy as np
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
df = pd.read_excel('RBD.xlsx')
df
```

```python
data = pd.melt(
    df.reset_index(),
    id_vars=['index'],
    value_vars=['System A', 'System B', 'System C']
)
```

```python
data.columns = ['index', 'treatments', 'value']
model = ols('value ~ C(treatments)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=1)
```

```python
anova_table
```

```python
# Accept the null hypothesis
```

```python
# SOlving RBD Example
data = pd.melt(
    df.reset_index(),
    id_vars=['index'],
    value_vars=['System A', 'System B', 'System C']
)
```

```python
data.columns = ['block', 'treatments', 'value']
model = ols('value ~ C(block) + C(treatments)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=1)
```

```python
anova_table
```

```python
# Reject the null hypothesis
```
