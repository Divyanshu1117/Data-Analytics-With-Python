```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_excel('reg2.xlsx')
```

```python
x = data['Hydrocarbon level']
y = data['O2']
```

```python
plt.figure()
sns.regplot(x=x, y=y, fit_reg=True)
```

```python
plt.scatter(np.mean(x), np.mean(y), color="green")
```

```python
plt.xlabel("Hydrocarbon level")
plt.ylabel("O2")
```

```python
plt.show()
```
