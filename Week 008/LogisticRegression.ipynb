```python
import statsmodels.api as sm
```

```python
import pandas as pd
```

```python
df = pd.read_excel("Simmons.xls")
```

```python
print(df.head())
```

```python
x = df[['Card', 'Spending']]
y = df['Coupon']
```

```python
df.columns
```

```python
print(df.columns)
```

```python
x1 = sm.add_constant(x)
```

```python
logit_model = sm.Logit(y, x1)
```

```python
result = logit_model.fit()
```

```python
print(result.summary())
```

```python
result.predict(x1)
```

```python
new_data = pd.DataFrame({
    'const':[1],
    'card':[1],
    'Spending':[2]
})
```

```python
result.predict(new_data)
```
