```python
def CDFExponential(lamb, x):  # lamb = lambda (rate parameter)
    if x <= 0:
        cdf = 0
    else:
        cdf = 1 - np.exp(-lamb * x)
    return cdf


# Example usage
lamb = 0.5
x = 2

result = CDFExponential(lamb, x)
print("CDF Value:", result)
```
