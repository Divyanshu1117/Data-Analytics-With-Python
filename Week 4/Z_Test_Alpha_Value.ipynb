```python
from scipy import stats
import numpy as np
```

```python
# Defining function for calculating alpha value
def z_value(x, mu, SEM):
    z = (x - mu) / SEM
    
    if z < 0:
        alfa = stats.norm.cdf(z)
    else:
        alfa = 1 - stats.norm.cdf(z)
    
    print(alfa)
```

```python
# Calculating alpha for different values of x, mu, and SEM
x = 48.5
mu = 50
SEM = 0.79
```

```python
z_value(x, mu, SEM)
```

```python
# Type 1 Error
z_value(48, mu, SEM)
```

```python
z_value(52, mu, SEM)
```

```python
# Type II Error (β / beta)
def type_2(mu1, mu2, sigma, n, alfa):
    z = stats.norm.ppf(alfa)
    xbar = mu1 + (z * sigma / np.sqrt(n))
    z2 = (xbar - mu2) / (sigma / np.sqrt(n))
    
    if mu1 > mu2:
        beta = 1 - stats.norm.cdf(z2)
    else:
        beta = stats.norm.cdf(z2)
    print(beta)
```

```python
# Function call
type_2(8.3, 7.4, 3.1, 60, 0.05)
```
