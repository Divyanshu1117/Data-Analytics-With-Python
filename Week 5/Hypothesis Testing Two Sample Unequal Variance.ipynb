```python
from scipy import stats
```

```python
# critical t value
critical_t = stats.t.ppf(0.025, 13)
print(critical_t)
```

```python
# sample data
metro = [3, 7, 25, 10, 15, 6, 12, 25, 15, 7]
rural = [48, 44, 40, 38, 33, 21, 20, 12, 1, 18]
```

```python
# two sample t-test (unequal variance)
result = stats.ttest_ind(metro, rural, equal_var=False)
print(result)
```

```python
# sample data
KARL = [1.186, 1.151, 1.322, 1.339, 1.200, 1.402, 1.365, 1.537, 1.559]
LEH  = [1.061, 0.992, 1.063, 1.062, 1.065, 1.178, 1.037, 1.086, 1.052]
```

```python
# paired t-test (dependent samples)
result = stats.ttest_rel(KARL, LEH)
print(result)
```
